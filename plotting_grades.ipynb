```python
from pandas import *
```

```python
# here insert the pair aluno:grade, as depicted below, and after, just run the code.
aluno_nota = {'Student_A':4.0, 'Student_B':6.0, 'Student_C':0.5}

alunos = list(aluno_nota)
notas = [aluno_nota[i] for i in alunos]
data = {'Student': alunos, 'Grade':notas}
df = DataFrame(data)
df
```

```python
from numpy import *
from matplotlib.pyplot import *


'''Len (grades)'''
N = len(notas)
area = 350

figure(figsize = (12, 12), facecolor = 'lightyellow').add_subplot(111, projection='polar')

for i in range(0,N):
    
    if (notas[i] < 5):
        r = notas[i]
        theta = 2 * pi * random.rand(N)
        scatter(theta[i], r, c='red', edgecolors = 'black', s=area, cmap='hsv', alpha=0.8)        
    else:
        r = notas[i]
        theta = 2 * pi * random.rand(N)
        scatter(theta[i], r, c='blue', edgecolors = 'black', s=area, cmap='hsv', alpha=0.8)

t = linspace(0, 2 * pi)
ycirc1 = [5 for i in t]
plot(t, ycirc1, '--', color = 'red')

ycirc2 = [df.mean() for i in t]
plot(t, ycirc2, '--', color = 'darkgreen')

fill_between(t, 0, ycirc1, facecolor = 'red', alpha = 0.1)
fill_between(t, 5, 11, facecolor = 'lightblue', alpha = 0.1)

legend(('Mean', 'Mean(class)'), loc = 'best', fontsize = 25)

title('Grades Distribution', size = 30, color = 'black')
xticks([])
yticks([2.0, 5.0, 10.0], fontsize = 20)
ylim(0, 11)
grid()
savefig('grades.png', facecolor = 'lightgreen')
show()
```
